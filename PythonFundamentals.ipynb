## Fibonacci Sequence 

- takes the past two outputs and sums them together

**Objective:** Our goal is to find what the Fibonacci number is in nth place

**Learned:**

- place the list in the function if you will be using appends or else the list infinitely stores the function outputs
- When a condition is found to be TRUE in an if statement it will exit the if block allowing external counters to exist
- how to use while loops to look up dynamically changing conditions

```python
import pandas as pd


def fibonacci_calculation(n):
        fib_list = []
        current_index = 0 
        
        # if statements skip entire blocks if the condition is met, that is how we can incrementat in the while loop 
        while n >= current_index:
            if current_index == 0:
                fib_list.append(0)
            elif current_index == 1:
                fib_list.append(1)
            else:
                second_closest_result = fib_list[-2]
                previous_result = fib_list[-1]
                fib_list.append(previous_result + second_closest_result)
            current_index += 1
        
        return fib_list

```

```python
n = 2
fibonacci_list = pd.DataFrame(fibonacci_calculation(n))
#fibonacci_list.to_csv('fibonacci_sequence.csv', index=False, header=False)

fibonacci_list
```

## Memory Efficent Fibonacci Calculation

Instead of reading through (and storing) n items in memory we instead just have 4 variables taking this close to O(1)


```python
def fibonacci_calculation_optimized(n):
    current_number = 0
    second_previous_result = 0 
    previous = 1

    current_index = 0
    while n >= current_index:
        if current_index == 0:
            current_number = second_previous_result
        elif current_index == 1:
             current_number = previous
        else:
            current_number = second_previous_result + previous
            second_previous_result = previous
            previous = current_number
        current_index += 1

        
    return current_number
```

```python
print(fibonacci_calculation_optimized(0))  # Should be 0
print(fibonacci_calculation_optimized(1))  # Should be 1
print(fibonacci_calculation_optimized(2))  # Should be 1
print(fibonacci_calculation_optimized(10)) # Should be 55
```
